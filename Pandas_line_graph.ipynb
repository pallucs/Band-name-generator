```python
import pandas as pd
```

```python
#CREATING TWO FUNCTIONS ON THE SAME GRAPH
# USING A DICTIONARY TO SHOW FUNCTIONS AS KEYS AND LIST OF NUMBERS AS VALUES
y = {'f(x)': [2,10,6,8], 'g(x)': [4,2,-2,6]}
x = [1,2,3,4]

graph = pd.DataFrame(y,x)
graph.plot(kind='line', grid=True, title='Graph', ylabel='Y title', xlabel='X title')
```
